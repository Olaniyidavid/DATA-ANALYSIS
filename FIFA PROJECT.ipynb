# CONCATENATE THE ALIKE COLUMNS INTO A SINGLE COLUMN e.g shortpassing and longpassing as passing

```python
attributes['Passing']  = attributes[['LongPassing','ShortPassing']].mean(axis=1)
```

```python
attributes['Pace'] = attributes[['Acceleration','SprintSpeed']].mean(axis=1)
```

```python
attributes['Physicality'] = attributes[['Stamina','Strength','Agility']].mean(axis=1)
```

```python
attributes['Tackling'] = attributes[['StandingTackle','SlidingTackle']].mean(axis=1)
```

```python
attributes['Shooting'] = attributes[['LongShots','ShotPower']].mean(axis=1)
```

# DROP THE FORMER COLUMNS

```python
attributes = attributes.drop(['Acceleration','SprintSpeed','Stamina','Strength','Agility','StandingTackle','SlidingTackle','LongShots','ShotPower','LongPassing','ShortPassing'],axis = 1)
```

4. Which countries have the highest overall ratings for their male players in FIFA 23?

```python
top_10 = cont_rating.loc[cont_rating['Overall'].nlargest(10).index]
```
