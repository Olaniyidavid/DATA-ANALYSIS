```python
#renaming the Best Position column
attributes.rename(columns={'Best Position':'Position'},inplace=True)
```

```python
#since CF and ST refer to the same position we replace CF with ST.
attributes['Position'].replace('CF','ST',inplace=True)
```

```python
#since LWB and LB refer to the same position we replace LWB with LB.
attributes['Position'].replace('LWB','LB',inplace=True)
```

```python
#since RWB and RB refer to the same position we replace RWB with RB.
attributes['Position'].replace('RWB','RB',inplace=True)
```

```python
dist_position = attributes['Position'].value_counts()
```

```python
dist_position.plot(kind = 'bar',color ='purple')
plt.title('Distribution of male player positions  in FIFA 23', y=1.12)
plt.xlabel("Position")
plt.ylabel("Count")
plt.show()
```

# CONCATENATE THE ALIKE COLUMNS INTO A SINGLE COLUMN e.g shortpassing and longpassing as passing

```python
attributes['Passing']  = attributes[['LongPassing','ShortPassing']].mean(axis=1)
```

```python
attributes['Pace'] = attributes[['Acceleration','SprintSpeed']].mean(axis=1)
```

```python
attributes['Physicality'] = attributes[['Stamina','Strength','Agility']].mean(axis=1)
```

```python
attributes['Tackling'] = attributes[['StandingTackle','SlidingTackle']].mean(axis=1)
```

```python
attributes['Shooting'] = attributes[['LongShots','ShotPower']].mean(axis=1)
```

# DROP THE FORMER COLUMNS

```python
attributes = attributes.drop(['Acceleration','SprintSpeed','Stamina','Strength','Agility','StandingTackle','SlidingTackle','LongShots','ShotPower','LongPassing','ShortPassing'],axis = 1)
```

# DATA ANALYSIS QUESTIONS ANSWERED WITH PYTHON

1. Who is the highest rated male player in FIFA 22, and what are his attributes?

```python
max_rating = attributes.loc[attributes['Overall'].idxmax()]
print(max_rating)
```

2. What is the average age of male players in FIFA 22?

```python
average_age = attributes['Age'].mean()
print(average_age)
```

3. What is the correlation between a player's overall rating and their individual attributes.

```python
attributes.corr()['Overall'].sort_values().to_frame()
```

4. Which countries have the highest overall ratings for their male players in FIFA 22?

```python
grp = attributes[['Overall','Nationality']]
cont_rating =grp.groupby(['Nationality'], as_index = False).mean()
cont_rating
```

```python
top_10 = cont_rating.loc[cont_rating['Overall'].nlargest(10).index]
```

```python
top_10
```
